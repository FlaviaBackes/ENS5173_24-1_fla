```python
import numpy as np

# Criando um vetor de dados aleatórios
data = np.random.rand(1000,2)

# Verificando o tamanho do vetor
print(data.shape)
print(np.size(data))
#shape do vetor com linhas e com colunas
#imprimiu todos os elementos com todo o tamanho do vetor, com 2 mil
# (mil linhas, 2 colunas)

# Print de uma amostra dos dados
print(data[0:10])
print(data[0:10].shape)
# imprimiu os primeiros 10 registros dos dados
# imprimiu shape da amostra
```

```python
# Usando Monte-Carlo - ajuste dos valores de data
dataMax = 1000
dataMin = 10
data = (dataMax - dataMin)*data + dataMin
print(data[0:10])
# multiplicou data pela subtração e intervalo entre data max e min
```

```python
import matplotlib.pyplot as plt #aqui foi necessário ativar no cmd essa biblioteca

fig, ax = plt.subplots(3) #grafico com 3 eixos
ax[0].plot(data) # dados do primeiro grafico
ax[1].plot(data[:,0]) # 1° dimensão de dados no grafico 2
ax[2].plot(data[:,1]) # 2° dimensão de dados no grafico 3
print(type(data)) #impressao do tipo de variavel
```

```python
# Criando uma função
def noiseCreation(path, year):
    """Esta função cria um ruido com sazonalidade.
    Inputs:
        path = asdasdas
        year = asdasdasdas

    Criado por Leonardo
    """
    #salvo em csv
    # Criando um dataframe
    df = pd.DataFrame()
    
    # Criando vertor temporal
    dtime = np.arange(np.datetime64(str(year)+"-01-01"), np.datetime64(str(year+1)+"-01-01"), np.timedelta64(1, "h"))
    
    # Monstrando um dataframe
    df['datetime'] = dtime
        
    # Datetime como index do dataframe
    df = df.set_index(df['datetime'])
    
    # Criando colunas com ano, mês e dia
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    
    # Crio uma coluna random
    df['random'] = np.random.rand(df.shape[0])
    
    # Adicionar sazonalidade
    tt = np.arange(0,df.shape[0],1)
    print(tt.shape)
    df['sazonal'] = (np.cos(tt/(360*4)))
    df['timeSeries'] = 10*df['sazonal']+df['random']
    #df['sazonal'].plot()
    df['timeSeries'].plot(color='red')
    df.to_csv(path+'/noise_'+str(year)+'.csv')
    
    return df

#path é o caminho pra salvar o arquivo csv
#year é o ano em que esta gerando o ruido sazonal
#dtime é o vetor temporal, com datas e horas no ano especifico
#df é retornado e plotado grafico da serie temporal
```

```python
import os
files = os.listdir(path)
print(files)

#lista arquivos no diretorio

allData = []
for file in files:
    if file.endswith('.csv'):
        print(file)
        data = pd.read_csv(path+'/'+file)
        allData.append(data)

#iteração de cada arquivo quando tem csv
#dados armazenados em df e na lista
```

```python
allDataDf = pd.concat(allData)
#concatena os dfs na lista alldata no eixo x
# vira unico df, com todos os dados
```
