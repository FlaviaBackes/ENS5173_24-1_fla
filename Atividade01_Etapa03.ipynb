```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dados originais
anos = np.array([2010, 2011, 2012, 2013, 2014])
producao = np.array([469000, 417000, 424000, 427000, 431000])

# Ajuste de uma linha de tendência linear
tendencia = np.polyfit(anos, producao, 1)

# Função para prever a produção com base no ano
def prever_producao(ano):
    return tendencia[0] * ano + tendencia[1]

# Anos para os quais queremos prever (até 2050)
anos_projetados = np.arange(2010, 2051)

# Previsões de produção para os próximos anos
producao_projetada = prever_producao(anos_projetados)

# Criar DataFrame com os dados projetados
dados_projetados = pd.DataFrame({'Ano': anos_projetados, 'Produção (toneladas)': producao_projetada})

# Exibir os dados projetados
print(dados_projetados)

# Plotar gráfico
plt.figure(figsize=(10, 6))
plt.plot(anos, producao, 'o-', label='Dados Originais')
plt.plot(anos_projetados, producao_projetada, 'r--', label='Projeção')
plt.title('Produção ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Produção (toneladas)')
plt.legend()
plt.grid(True)
plt.show()
```
