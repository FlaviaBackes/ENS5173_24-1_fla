```python
# Criando uma Lista em Python
list = [0, 'amo', 2.5]
print(list)

# Nesta parte fiz a adição entre duas strings, sendo que list[1] é o segundo indice do vetor list
print('variável: '+list[1])

# Adição do primeiro indice da variavel list (list[0]) - Lembrar que indexação inicia em zero
print(list[0] + 1)
```

```python
# Criando uma matriz numpy
import numpy as np

#criando matriz de zeros utilizando np.zeros
#se eu colocar np. e der uma tab, encontro varias funções e como declarar
a= np.zeros((5,5))
print(a)

print('indexação em python')
print(a[0,0])
```

```python
#Criando n°s aleatórios e mostrando a dimensão da matriz
b = np.random.rand(5,5)
print(b)

#print da dimensão
print('Dimensão da Matriz')
print(b.shape)
```

```python
# Criando Matriz de valores aleatorios entre uma faixa estabelecida
vmin = 1
vmax = 100

c= vmin+(vmax-vmin)*b
print(c)
```

```python
# Plotar matriz
import matplotlib.pyplot as plt

plt.plot(c.transpose())
```

```python
#Plotando apenas a primeira linha

x=np.array(range(0,np.size(c[0,:])))
print(x)
plt.plot(x,c[0,:])
plt.title('Gráfico com Valores Inteiros no Eixo X')
plt.xticks(x)
```

```python
#Plotando apenas a primeira linha

x=np.array(range(0,np.size(c[0,:])))
print(x)
plt.scatter(x,c[0,:])
```

```python
# Criando um graf de barras e exportando

fig, ax=plt.subplots(3,1)  #plota ax pq é um objeto plt
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

# exportar
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder=rootFolder.replace('\\','/')+'/figures/'
print(figFolder)

fig.savefig('Figure01.png')
```
